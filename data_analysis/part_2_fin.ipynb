### Задание 1 (0.5 балла)

Создайте массив Numpy под названием `a` размером 5x2, то есть состоящий из 5 строк и 2 столбцов. Первый столбец должен содержать числа `1, 2, 3, 3, 1`, а второй - числа `6, 8, 11, 10, 7`.  
Будем считать, что каждый столбец - это признак, а строка - наблюдение. Затем найдите среднее значение по каждому признаку, используя метод `mean` массива Numpy.  
Результат запишите в массив `mean_a`, в нем должно быть 2 элемента.

````python
import numpy as np
a=np.array([
      [1,6],
      [2,8],
      [3,11],
      [3,10],
      [1,7]])
mean_a=np.mean(a, axis=0)
print(mean_a)
````

### Задание 2 (0.5 балла)

Вычислите массив `a_centered`, отняв от значений массива `а` средние значения соответствующих признаков, содержащиеся в массиве `mean_a`.  
Вычисление должно производиться в одно действие. Получившийся массив должен иметь размер 5x2.

````python
a_centered=a-mean_a
print(a_centered)
````

### Задание 3 (0.5 балла)

Найдите скалярное произведение столбцов массива `a_centered`. В результате должна получиться величина `a_centered_sp`.  
Затем поделите `a_centered_sp` на `N-1`, где `N` - число наблюдений.

````python
a_centered_sp=np.dot(a_centered[:,0],a_centered[:,1])
N=a.shape[0]
cov=a_centered_sp/(N-1)
print(cov)
````

Число, которое мы получили в конце задания 3 является ковариацией двух признаков, содержащихся в массиве `а`. В задании 3 мы делили сумму произведений центрированных признаков на `N-1`, а не на `N`, поэтому полученная нами величина является несмещенной оценкой ковариации.  

В этом задании проверьте получившееся число, вычислив ковариацию еще одним способом - с помощью функции `np.cov`. В качестве аргумента `m` функция `np.cov` должна принимать транспонированный массив `a`. В получившейся ковариационной матрице (массив Numpy размером 2x2) искомое значение ковариации будет равно элементу в строке с индексом 0 и столбце с индексом 1.


Примечание:

Ковариация — это мера линейной зависимости между двумя случайными величинами. Она показывает, как переменные изменяются вместе: если при увеличении одной переменной другая также имеет тенденцию к увеличению, ковариация положительна; если при увеличении одной другой имеет тенденцию к уменьшению — отрицательна. Если ковариация равна нулю, это означает отсутствие линейной зависимости между переменными.

Ковариация также является основой для более сложных статистических понятий, таких как корреляция, которая нормализует ковариацию, делая её независимой от масштабов измерений. Это позволяет более наглядно оценивать степень связи между переменными. Важно отметить, что ковариация не указывает на причинно-следственную связь, а лишь показывает, как две величины могут изменяться одновременно. Таким образом, ковариация является мощным инструментом для анализа взаимосвязей в данных и играет ключевую роль в статистическом моделировании и машинном обучении.

````python
cov_matrix=np.cov(a.T)
cov_value=cov_matrix[0,1]
print(cov_value)
````

Создайте датафрейм `authors` со столбцами `author_id` и `author_name`, в которых соответственно содержатся данные:  
`[1, 2, 3],  
['Тургенев', 'Чехов', 'Островский']`.

Затем создайте датафрейм `books` cо столбцами `author_id`, `book_title` и `price`, в которых соответственно содержатся данные:  
`[1, 1, 1, 2, 2, 3, 3],  
['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],  
[450, 300, 350, 500, 450, 370, 290]`.


````python
import pandas as pd
authors=pd.DataFrame({'author_id':[1,2,3],
                      'author_name':['Тургенев','Чехов','Островский']})
books=pd.DataFrame({'author_id':[1,1,1,2,2,3,3],
                     'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                     'price':[450, 300, 350, 500, 450, 370, 290]})
authors.index = range(1, len(authors) + 1)
books.index = range(1, len(books) + 1)

print(authors)
print(books)
````

### Задание 6 (0.5 балла)

Получите датафрейм `authors_price`, соединив датафреймы `authors` и `books` по полю `author_id`.

````python
authors_price=pd.merge(authors,books, on= 'author_id')
authors_price.index = range(1, len(authors_price) + 1)
print(authors_price)
````

### Задание 7 (0.5 балла)

Создайте датафрейм `top5`, в котором содержатся строки из `authors_price` с пятью самыми дорогими книгами.

````python
top5 = authors_price.sort_values(by='price',ascending=False).head(5)
print(top5)
````

### Задание 8 (0.5 балла)

Создайте датафрейм `authors_stat` на основе информации из `authors_price`.  
В датафрейме `authors_stat` должны быть четыре столбца:
`author_name`, `min_price`, `max_price` и `mean_price`,
в которых должны содержаться соответственно имя автора, минимальная, максимальная и средняя цена на книги этого автора.


````python
authors_stat = authors_price.groupby('author_name')['price'].agg(min_price='min',max_price='max',mean_price='mean').reset_index()
authors_stat.index = range(1, len(authors_stat) + 1)
print(authors_stat)
````

Используйте для линий разные цвета, одна из них пусть будет пунктирной. Результат в этом и следующих заданиях на построение графиков должен получиться максимально схожим с картинкой. Содержание подписей не должно обязательно совпадать, однако осмысленные подписи должны присутствовать на графиках.

````python
import numpy as np
import matplotlib.pyplot as plt

# Создание массива x от -π до π 100 точек
x = np.linspace(-np.pi, np.pi, 100)

# функции
y_sin = np.sin(x)
y_cos = np.cos(x)

# График (размер, цвет,форма)
fig, ax = plt.subplots(figsize=(4.3, 2.9))
fig.patch.set_facecolor('white')
ax.plot(x, y_sin, color='mediumturquoise', label='sin(x)',linewidth=1.3)
ax.plot(x, y_cos, color='darkmagenta', linestyle='--', label='cos(x)',linewidth=1.3)

ax.axhline(0,color="gray",linewidth=0.3)
ax.axvline(0,color="gray",linewidth=0.3)
ax.grid(True,linestyle='-',alpha=0.3,linewidth=0.3,color='gray')


#настройка осей (шаг, размер)
ax.set_xticks(np.arange(-3, 4, 1))
ax.tick_params(axis='x', labelsize=6.4)
ax.set_yticks(np.arange(-1, 1.25, 0.25))
ax.tick_params(axis='y', labelsize=6.4)


#название графика
ax.set_title('Функции sin(x) и cos(x)', fontsize=7)

# Цвет фона
ax.set_facecolor('white')

# Настройка рамки
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_color('black')
    spine.set_linewidth(0.5)

#Настройка легенды (размер,положение, цвет фона)
legend = ax.legend(fontsize=6.3, loc='upper left')
legend.get_frame().set_facecolor('white')
legend.get_frame().set_edgecolor('gray')

plt.show()
````

Прочитайте средствами pandas файл с данными кредитного скоринга. Рассмотрите сравнительное количество положительных и отрицательных кейсов в зависимости от размера текущего счёта. Постройте зеркальные горизонтальные гистограммы.

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(GERMAN_CREDIT)

crosstab = pd.crosstab(df['status'], df['credit_risk'])

categories = crosstab.index.tolist()
good_counts = crosstab['good'].values  # или crosstab.iloc[:, 0]
bad_counts = crosstab['bad'].values    # или crosstab.iloc[:, 1]

y= np.arange(len(categories))

max_x = max(max(good_counts), max(bad_counts))

# Создание двух отдельных графиков
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))


fig.suptitle('Status of checking account', fontsize=9.5)

# левая
ax1.set_facecolor('white')
ax1.barh(y, good_counts, label='Good risk', color='green', height=0.5)
ax1.set_yticks(y)
ax1.grid(False)
ax1.set_yticklabels(categories, fontsize=8.5)
ax1.set_title('Good risk', fontsize=9.5)
ax1.set_xticks(np.arange(0, max_x + 50, 50))
ax1.invert_xaxis()

for spine in ax1.spines.values():
    spine.set_visible(True)
    spine.set_color('black')
    spine.set_linewidth(0.5)

# правая
ax2.set_facecolor('white')
ax2.barh(y, bad_counts, label='Bad risk', color='red', height=0.5)
ax2.set_yticks(y)
ax2.grid(False)
ax2.set_yticklabels(categories, fontsize=8.5)
ax2.set_title('Bad risk', fontsize=9.5)
ax2.set_xticks(np.arange(0, max_x + 50, 50))

for spine in ax2.spines.values():
    spine.set_visible(True)
    spine.set_color('black')
    spine.set_linewidth(0.5)


ax2.tick_params(labelright=True, right=True, labelleft=False, left=False) #подписи справа

plt.tight_layout()
plt.show()
````
